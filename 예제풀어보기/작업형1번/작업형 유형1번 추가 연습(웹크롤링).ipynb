# 4월 1일 부터 4월 30일까지 영화 랭킹 조회

## url 생성 함수 작성

```python
# print("{:02d}".format(9))  : formatting
## url을 이용하여 웹 크롤링 하는 함수 작성
def get_searchUrl(day:int):
    baseurl="https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=202104"
    searchurl=baseurl+"{:02d}".format(day)
    return searchurl

```

```python
get_searchUrl(12)[-2:]
```

```python
## title , 평점, 링크, date 리스트로 생성하는 함수 작성
```

```python
def get_movieData(searchurl:str):
    url=searchurl
    data=requests.get(url)
    soup=bs(data.text)
    lstRanking=soup.find(class_="list_ranking") #class 찾기
    title=[x.text.strip() for x in lstRanking.find_all("div",class_='tit5')]
    point=[x.text for x in lstRanking.find_all("td",class_="point")]
    rank=range(1,len(point)+1)
    movieTable=pd.DataFrame({'rank':rank,"title":title,"point":point,"link":link})
    movieTable['date']=pd.to_datetime("2021-04-"+searchurl[-2:])
    return movieTable


```

```python
df_list=[] # dataframe 넣을 리스트 선언
for day in range(1,31):
    df=get_movieData(get_searchUrl(day))
    df_list.append(df)
```

## 결과 확인하기

```python
total=pd.concat(df_list).reset_index(drop=True)
total.head()
```
